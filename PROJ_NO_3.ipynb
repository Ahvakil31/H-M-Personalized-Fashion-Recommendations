## Label Encoding for Categorical Features

```python
from pyspark.ml.feature import StringIndexer

# Initialize StringIndexer for 'prod_name'
indexer = StringIndexer(inputCol="prod_name", outputCol="prod_name_indexed")

# Fit the indexer to the DataFrame and transform the column
df_encoded = indexer.fit(df).transform(df)

# Display the schema and the first few rows to verify the new column
df_encoded.printSchema()
df_encoded.select("prod_name", "prod_name_indexed").show(1000, truncate=False)
```

```python
from pyspark.sql.functions import col

# Count the number of unique product names
unique_prod_names_count = df_encoded.select('prod_name_indexed').distinct().count()
print(f"Total number of unique product names: {unique_prod_names_count}")

# Count the occurrences of each product name
prod_name_counts = df_encoded.groupBy('prod_name_indexed').count().orderBy(col('count').desc())

print("\nTop 10 product names and their counts:")
prod_name_counts.show(100000,truncate=False)
```

```python
from pyspark.ml.feature import StringIndexer
from pyspark.sql import SparkSession

# Ensure SparkSession and SparkContext are active
spark_session_re_check = SparkSession.builder.getOrCreate()

# Define the categorical columns to encode
categorical_cols_to_encode = ['product_type_name', 'index_name']

# Apply StringIndexer to each categorical column
current_df = df_encoded # Start with the already encoded 'prod_name'

for col_name in categorical_cols_to_encode:
    indexer = StringIndexer(inputCol=col_name, outputCol=col_name + "_indexed")
    current_df = indexer.fit(current_df).transform(current_df)

df_encoded_all = current_df

# Display the schema to verify the new columns
df_encoded_all.printSchema()

# Display a sample of the original and indexed columns for verification
df_encoded_all.select(
    'product_type_name', 'product_type_name_indexed',
    'index_name', 'index_name_indexed'
).show(5, truncate=False)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation

# Identify numerical and indexed columns for correlation analysis
numerical_cols = [
    'article_id', 'product_code', 'product_type_no', 'graphical_appearance_no',
    'colour_group_code', 'perceived_colour_value_id', 'perceived_colour_master_id',
    'department_no', 'index_group_no', 'section_no', 'garment_group_no'
]
indexed_cols = [
    'prod_name_indexed', 'product_type_name_indexed', 'index_name_indexed'
]

# Combine all relevant columns
all_features = numerical_cols + indexed_cols

# Create a VectorAssembler to combine features into a single vector column
assembler = VectorAssembler(
    inputCols=all_features,
    outputCol="features"
)

# Transform the DataFrame to include the features vector
df_features = assembler.transform(df_encoded_all)

# Calculate the correlation matrix
# Using Pearson correlation (default)
correlation_matrix = Correlation.corr(df_features, "features").head()

# Convert the correlation matrix to a dense matrix and print it
correlation_array = correlation_matrix[0].toArray()

import pandas as pd
correlation_df = pd.DataFrame(correlation_array, index=all_features, columns=all_features)

print("Correlation Matrix:")
print(correlation_df)
```
