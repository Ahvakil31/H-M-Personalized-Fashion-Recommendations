```python
from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("CSV_Reader") \
    .getOrCreate()

df = spark.read.csv("/content/drive/MyDrive/ZAALIMA PROJECT/DATASET PROJECT MAY/articles_hm.csv", header=True, inferSchema=True)
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
df.count()
```

```python
from pyspark.sql.functions import col

# Count the number of unique product names
unique_prod_names_count = df.select('prod_name').distinct().count()
print(f"Total number of unique product names: {unique_prod_names_count}")

# Count the occurrences of each product name
prod_name_counts = df.groupBy('prod_name').count().orderBy(col('count').desc())

print("\nTop 10 product names and their counts:")
prod_name_counts.show(10, truncate=False)
```

```python
from pyspark.sql.functions import col, sum

# Check for null values in each column
null_counts = df.select([sum(col(c).isNull().cast('integer')).alias(c) for c in df.columns])

print("Number of null values per column:")
null_counts.show(truncate=False)
```

## Drop detail_desc column

### Subtask:
Drop the 'detail_desc' column from the DataFrame.


**Reasoning**:
I need to drop the 'detail_desc' column from the DataFrame `df` and then display the schema to confirm the column has been dropped.



```python
df = df.drop('detail_desc')
df.printSchema()
```

```python
from pyspark.sql.functions import col, sum

# Check for null values in each column
null_counts = df.select([sum(col(c).isNull().cast('integer')).alias(c) for c in df.columns])

print("Number of null values per column:")
null_counts.show(truncate=False)
```
